My decision rules:

**Step 3: Implement and Test** (10 minutes)

Now write the Python code for your system. You can:

**Code your solution here:**

````python
# Your study recommendation system
# GPA Input (Enter whole numbers: 0, 1, 2, 3, 4, 5)
gpa = float(input("Enter your GPA (0-5): "))

# 1. Highest Priority: Perfect Score
if gpa == 5:
    print("Status: Perfect Score")

# 2. Range 3 to 4: Pass
elif gpa >= 3:
    print("Status: Pass")

# 3. Range 2: Low Pass
elif gpa >= 2:
    print("Status: Low Pass")

# 4. Everything else (0 to 1): Fail
else:
    print("Status: Fail")
    # Personalized Study Advice System

# Step 1: Check the Grade
grade = int(input("Enter your grade (0-5): "))

if grade > 4:
    # No intervention for Perfect Scores
    print("Excellent! You are doing a great job. Keep it up!")

else:
    # Step 2: Intervention - Gather more data
    print("Let's look at your habits to help improve that grade.")
    sleep = int(input("How many hours of sleep did you get? "))
    hours_studied = int(input("How many hours have you studied today? "))

    print("\n--- Recommendation ---")

    # Priority 1: Health (Sleep)
    if sleep < 8:
        print("Advice: Try to get more sleep. Your brain needs 8+ hours to process what you learn.")

    # Priority 2: Overworking (Burnout)
    elif hours_studied > 8:
        print("Advice: Take a rest! Studying over 8 hours can lead to burnout and lower retention.")

    # Priority 3: Underworking
    elif hours_studied < 2:
        print("Advice: Try to study more. Aim for at least 3-4 hours of focused work.")

    # Priority 4: Balanced
    else:
        print("Advice: You have a healthy balance. Keep this consistency to raise your grade!")

````

## Key Takeaways

🧠 **Computational Thinking**:

- Always understand the problem before coding
- Break complex decisions into clear rules
- Test your logic with examples before implementing

🤖 **AI Partnership**:

- AI works best when you have clear requirements
- Use AI to implement your ideas, not to think for you
- Ask AI to explain concepts you don't understand

🎯 **Decision Patterns**:

- if/else for two choices
- if/elif/else for multiple categories
- Order matters in elif chains
