## Part 2: Financial Decision Logic (15 minutes)

### Budget Limit Checker
Apply your if/else skills to real budget management:

````python
print("=== Budget Limit Checker ===")

# Get user's budget and expense
monthly_budget = float(input("Enter your monthly budget: $"))
expense = float(input("Enter expense amount: $"))

# Decision logic here - you write this part!
if expense > 0.5 * monthly_budget:
    print("Major purchase - think carefully")
    print("This is a large portion of your budget. Make sure it's a priority.")
elif expense > 0.25 * monthly_budget:
    print("Significant expense")
    print("This will impact your remaining balance. Plan accordingly.")
elif expense > 0.1 * monthly_budget:
    print("Moderate expense")
    print("A standard expense. Just be mindful of your total spending.")
else:
    print("Minor expense - within budget")
    print("This fits easily into your current plan")
````

### Savings Goal Progress
Use multiple conditions to track savings progress:

````python

print("=== Savings Goal Tracker ===")
#1. Get user input
goal_amount = float(input("Savings goal: $"))
current_savings = float(input("Current savings: $"))
#2. Calculate the percentage
if goal_amount > 0:
        percent = (current_savings / goal_amount) * 100
        print(f"\nProgress: {percent:.1f}%")

# 3. Categorize progress and provide feedback
        if percent >= 100:
            print("Status: Goal achieved! Consider a new goal.")
        elif percent >= 75:
            print("Status: Almost there! You're doing great.")
        elif percent >= 50:
            print("Status: Halfway there - keep it up!")
        elif percent >= 25:
            print("Status: Good progress, stay consistent.")
        else:
            print("Status: Just getting started - you've got this!")
````

## Part 4: Building Your Finance Tracker Foundation (15 minutes)

### Start Your Semester Project

Create a new notebook called "Personal_Finance_Tracker" and build the foundation:

````python
print("=== Personal Finance Tracker v0.1 ===")
print("Week 3: Smart Budget Decisions")
print()

# User setup
name = input("Enter your name: ")
monthly_income = float(input("Enter your monthly income: $"))

print(f"\nWelcome to your finance tracker, {name}!")
print(f"Monthly income: ${monthly_income:.2f}")

# This week's feature: Expense categorisation
print("\n--- Expense Entry ---")
expense_amount = float(input("Enter an expense: $"))
expense_description = input("What was this expense for? ")

# Expense categorisation logic
description = expense_description.lower()

if description in ["food", "restaurant", "groceries", "snacks"]:
    category = "Food"
elif description in ["rent", "electricity", "water", "internet"]:
    category = "Housing & Bills"
elif description in ["bus", "taxi", "uber", "didi", "transport"]:
    category = "Transportation"
elif description in ["movie", "games", "shopping", "entertainment"]:
    category = "Entertainment"
else:
    category = "Other"

print(f"\nCategory: {category}")

# Budget percentage calculation
expense_percentage = (expense_amount / monthly_income) * 100

print(f"This expense is {expense_percentage:.2f}% of your monthly income.")

# Helpful feedback
print("\n--- Budget Feedback ---")

if expense_percentage < 5:
    print("Good job! This expense is a small part of your budget.")
elif expense_percentage < 20:
    print("This is a moderate expense. Make sure it fits your monthly plan.")
else:
    print("Warning: This expense is quite large compared to your income.")
    print("Consider reviewing your spending habits.")

print("\nThank you for using Personal Finance Tracker!")
````
