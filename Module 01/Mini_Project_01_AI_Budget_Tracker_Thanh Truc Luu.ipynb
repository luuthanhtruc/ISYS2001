### Paste and Test Your AI-Generated Budget Tracker

```python
# AI-GENERATED BUDGET TRACKER CODE
# Ask for monthly income
income = input("Enter your monthly income: ")

# Ask for 3 expenses
expense1 = input("Enter expense 1: ")
expense2 = input("Enter expense 2: ")
expense3 = input("Enter expense 3: ")

# Convert inputs to numbers and calculate remaining budget
remaining = int(income) - (int(expense1) + int(expense2) + int(expense3))

# Show remaining budget
print("Remaining budget:", remaining)

# Show warning if budget is negative
if remaining < 0:
    print("⚠️ Warning: Your budget is negative!")
```

Choose ONE feature to add using AI assistance:

- 📊 Show percentage of income spent
- ✅ Input validation (no negative numbers)

```python
# Document your enhancement
enhancement_choice = "I chose to add: Input validation"
enhancement_prompt = """
[Write the prompt you used to add this feature]
"""

print(f"🚀 ENHANCEMENT: {enhancement_choice}")
```

```python
# ENHANCED BUDGET TRACKER
# Function to get a positive number from user
def get_positive_input(prompt):
    while True:
        value = input(prompt)
        if value.isdigit() and int(value) >= 0:
            return int(value)
        else:
            print("⚠️ Please enter a non-negative number!")

# Ask for monthly income
income = get_positive_input("Enter your monthly income: ")

# Ask for 3 expenses
expense1 = get_positive_input("Enter expense 1: ")
expense2 = get_positive_input("Enter expense 2: ")
expense3 = get_positive_input("Enter expense 3: ")

# Calculate remaining budget
remaining = income - (expense1 + expense2 + expense3)

# Show remaining budget
print("Remaining budget:", remaining)

# Show warning if budget is negative
if remaining < 0:
    print("⚠️ Warning: Your budget is negative!")
```
