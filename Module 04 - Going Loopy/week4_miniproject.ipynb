## Part 1: Enhanced Expense Entry (15 minutes)

### 🔄 From One to Many

Transform your Week 2 single-expense tracker to handle multiple expenses with proper validation:

```python
print("=== Enhanced Finance Tracker v0.3 ===")
print("Week 3: Loop-Powered Edition")
print()

# User setup (enhanced from Week 2)
name = input("Enter your name: ")

# Add budget validation using while loop
monthly_budget = 0
while monthly_budget < 100 or monthly_budget > 10000:
    try:
        budget_input = input("Enter monthly budget ($100-$10,000): $")
        monthly_budget = float(budget_input)

        if monthly_budget < 100:
            print("Budget too low! Minimum $100")
        elif monthly_budget > 10000:
            print("Budget seems high! Maximum $10,000")
    except ValueError:
        print("Please enter a valid number!")

print(f"\nWelcome {name}! Monthly budget: ${monthly_budget:.2f}")

# ========== MULTI-EXPENSE ENTRY ==========
expenses = []  # Store all expense amounts
descriptions = []  # Store what each expense was for
total_spent = 0

print("\n--- Enter Your Expenses ---")
print("Type 'done' when finished entering expenses")

# Your task: Complete the multi-expense entry system
# Use sentinel value pattern to collect multiple expenses
# Validate each expense amount (positive numbers only)
# Store both amount and description
# Show running total and budget percentage after each entry

while True:
    expense_input = input("\nExpense amount (or 'done'): $")

    if expense_input.lower() == 'done':
        break

    # Your validation and storage code here
    # The inner while True and duplicated expense_input were removed
    try:
        amount = float(expense_input)
        if amount <= 0:
            print("Amount must be positive!")
            continue
    except ValueError:
        print("Please enter a valid number!")
        continue

    desc = input("Description: ").strip() or "Unspecified"
    expenses.append(amount)
    descriptions.append(desc)
    total_spent += amount

    pct = (total_spent / monthly_budget) * 100
    print(f"✓ Added: {desc} — ${amount:.2f}")
    print(f"  Total: ${total_spent:.2f} ({pct:.1f}% of budget)")


print(f"\n📊 Expense entry complete: {len(expenses)} expenses, ${total_spent:.2f} total")
```

## Part 2: Basic Expense Analysis (15 minutes)

### 📈 Processing Your Transaction Data

Use loops to analyze the expenses you collected:

```python
# Expense Analysis Section
if len(expenses) > 0:
    print("\n📊 Expense Analysis:")

    # Basic statistics using loop patterns from workshops
    print(f"Total expenses: {len(expenses)}")
    print(f"Total spent: ${total_spent:.2f}")
    print(f"Average expense: ${total_spent/len(expenses):.2f}")
    print(f"Budget used: {(total_spent/monthly_budget)*100:.1f}%")
    print(f"Remaining budget: ${monthly_budget - total_spent:.2f}")

    # Your task: Find highest and lowest expenses
    # Use the min/max pattern from workshops
    highest_expense = max(expenses)
    lowest_expense  = min(expenses)

    print(f"\n💰 Expense Range:")
    print(f"Highest expense: ${highest_expense:.2f}")
    print(f"Lowest expense: ${lowest_expense:.2f}")

    # Categorize
    large_count = medium_count = small_count = 0
    for amt in expenses:
        pct = (amt / monthly_budget) * 100
        if pct > 20:
            large_count += 1
        elif pct >= 5:
            medium_count += 1
        else:
            small_count += 1

    print(f"\n📊 Expense Categories:")
    print(f"Large expenses (>20% budget): {large_count}")
    print(f"Medium expenses (5-20% budget): {medium_count}")
    print(f"Small expenses (<5% budget): {small_count}")

    # Smart feedback
    budget_pct = (total_spent / monthly_budget) * 100
    print("\n💡 Smart Feedback:")
    if budget_pct > 100:
        print("⚠️  You're over budget! Cut back on large expenses.")
    elif budget_pct > 80:
        print("🔶 Getting close to your limit — spend carefully.")
    elif budget_pct > 50:
        print("🟡 Halfway through your budget. You're doing okay.")
    else:
        print("✅ Great job! You're well within your budget.")

    if large_count >= 2:
        print(f"💸 You have {large_count} large expenses — consider if any can be reduced.")
    if small_count == len(expenses):
        print("👍 All expenses are small — very disciplined spending!")

else:
    print("\n📝 No expenses entered - try adding some transactions!")
```

## Part 3: Multiple Savings Goals Tracker (15 minutes)

### 🎯 Goal Management with Loops

Extend your tracker to handle multiple savings goals:

```python
print("\n=== Savings Goals Tracker ===")

# Goal collection
goal_names = []
goal_targets = []
goal_current = []

# Ask how many goals they want to track
num_goals = 0
while num_goals < 1 or num_goals > 5:  # Reasonable limits
    try:
        num_goals = int(input("How many savings goals? (1-5): "))
        if num_goals < 1:
            print("Must have at least 1 goal!")
        elif num_goals > 5:
            print("Maximum 5 goals for this tracker!")
    except ValueError:
        print("Please enter a valid number!")

# Your task: Collect goal information using for loop
# For each goal (1 to num_goals):
#   - Get goal name
#   - Get target amount (with validation)
#   - Get current amount (with validation)
#   - Store in the three lists

print(f"\nEntering {num_goals} savings goals:")

for i in range(num_goals):
    print(f"\n--- Goal {i+1} ---")

    # Your goal collection code here
for i in range(num_goals):
    print(f"\n--- Goal {i+1} ---")

    # Get goal name
    name = input("Enter goal name: ")
    goal_names.append(name)

    # Get target amount with validation
    while True:
        try:
            target = float(input("Enter target amount ($): "))
            if target <= 0:
                print("Target must be greater than 0!")
            else:
                goal_targets.append(target)
                break
        except ValueError:
            print("Please enter a valid number!")

    # Get current amount with validation
    while True:
        try:
            current = float(input("Enter current savings ($): "))
            if current < 0:
                print("Current savings cannot be negative!")
            else:
                goal_current.append(current)
                break
        except ValueError:
            print("Please enter a valid number!")

print(f"\n✅ Saved {len(goal_names)} goals!")
```

```python
# Goal Analysis and Progress Tracking
MONTHLY_SAVING = 100 # Define a default saving amount per month for calculation

if len(goal_names) > 0:
    print("\n📊 Goals Progress Analysis:")

    # Your task: Analyze each goal
    # For each goal, calculate:
    #   - Progress percentage
    #   - Amount remaining
    #   - Months to completion (if saving $100/month)

    total_target = 0
    total_current = 0
    progress_percentages = [] # Initialize the list to store progress percentages

    for i in range(len(goal_names)):
        # Your progress calculation here
        progress = (goal_current[i] / goal_targets[i]) * 100 if goal_targets[i] > 0 else 0
        remaining = goal_targets[i] - goal_current[i]
        months_to_completion = remaining / MONTHLY_SAVING if remaining > 0 else 0

        progress_percentages.append(progress)

        print(f"\n🎯 {goal_names[i]}")
        print(f"   Progress:     {progress:.1f}% (${goal_current[i]:.2f} / ${goal_targets[i]:.2f})")
        print(f"   Remaining:    ${remaining:.2f}")

        if remaining == 0:
            print(f"   Timeline:     ✅ Goal reached!")
        else:
            print(f"   Timeline:     ~{months_to_completion:.1f} months at ${MONTHLY_SAVING}/month")

        total_target += goal_targets[i]
        total_current += goal_current[i]

    # Overall progress summary
    overall_progress = (total_current / total_target) * 100 if total_target > 0 else 0
    total_remaining = total_target - total_current

    print(f"\n🎯 Overall Goals Summary:")
    print(f"Total target: ${total_target:.2f}")
    print(f"Total saved: ${total_current:.2f}")
    print(f"Overall progress: {overall_progress:.1f}%")
    print(f"Total remaining: ${total_remaining:.2f}")

    # Your task: Find which goal needs most attention
    # Find the goal with lowest progress percentage
    # Provide encouragement and specific advice
    lowest_index = 0
    for i in range(1, len(progress_percentages)):
        if progress_percentages[i] < progress_percentages[lowest_index]:
            lowest_index = i

    lowest_progress = progress_percentages[lowest_index]
    lowest_remaining = goal_targets[lowest_index] - goal_current[lowest_index]
    months_needed = lowest_remaining / MONTHLY_SAVING if lowest_remaining > 0 else 0

    print(f"\n💡 Focus Goal: '{goal_names[lowest_index]}'")

    if lowest_progress == 0:
        print(f"   You haven't started this one yet — even a small deposit gets the momentum going!")
    elif lowest_progress < 25:
        print(f"   Just {lowest_progress:.1f}% there. At ${MONTHLY_SAVING}/month you'd reach it in {months_needed:.1f} months.")
    elif lowest_progress < 75:
        print(f"   Good start at {lowest_progress:.1f}%! ${lowest_remaining:.2f} to go — you're past the hard part.")
    else:
        print(f"   So close at {lowest_progress:.1f}%! Only ${lowest_remaining:.2f} left — consider a one-time boost to finish it off.")

# This else statement correctly corresponds to the initial if len(goal_names) > 0:
else:
    print("No goals to analyze!")
```

## Part 4: Simple Menu System (10 minutes)

### 🎮 Interactive Program Flow

Create a basic menu that lets users choose what to do:

```python
print("\n=== Interactive Finance Menu ===")

# Simple menu system
running = True

while running:
    print("\n" + "-"*30)
    print(f"💰 {name}'s Finance Tracker")
    if len(expenses) > 0:
        # Ensure total_spent is updated if new expenses were added in menu option 3
        current_total_spent_display = sum(expenses)
        print(f"📊 Current: {len(expenses)} expenses, ${current_total_spent_display:.2f} spent")
    print("-"*30)
    print("1. View expense summary")
    print("2. View goals progress")
    print("3. Add new expense")
    print("4. Budget check")
    print("5. Exit")

    choice = input("\nChoose option (1-5): ")

    if choice == "1":
        print("\n--- Expense Summary ---")
        if len(expenses) == 0:
            print("No expenses recorded yet.")
        else:
            for i in range(len(expenses)):
                # Use 'descriptions' instead of 'expense_names'
                print(f"{i+1}. ${expenses[i]:.2f} - {descriptions[i]}")

            print(f"\nTotal spent: ${sum(expenses):.2f}")

    elif choice == "2":
        print("\n--- Goals Progress ---")
        if len(goal_names) == 0:
            print("No savings goals yet.")
        else:
            for i in range(len(goal_names)):
                percent = (goal_current[i] / goal_targets[i]) * 100 if goal_targets[i] > 0 else 0
                print(f"{goal_names[i]}: ${goal_current[i]:.2f}/${goal_targets[i]:.2f} ({percent:.1f}% complete)")

    elif choice == "3":
        print("\n--- Add New Expense ---")

        new_expense_desc = input("Enter expense description: ")

        while True:
            try:
                new_expense_amount = float(input("Enter expense amount: $"))
                if new_expense_amount <= 0:
                    print("Amount must be greater than 0!")
                else:
                    expenses.append(new_expense_amount)
                    descriptions.append(new_expense_desc)
                    global total_spent # Declare total_spent as global to modify it
                    total_spent += new_expense_amount # Update global total_spent
                    print("✅ Expense added successfully!")
                    break
            except ValueError:
                print("Please enter a valid number!")

    elif choice == "4":
        print("\n--- Budget Check ---")
        if monthly_budget > 0:
            current_total_spent = sum(expenses)
            budget_used_pct = (current_total_spent / monthly_budget) * 100
            remaining_budget = monthly_budget - current_total_spent
            print(f"Budget used: {budget_used_pct:.1f}%")
            print(f"Remaining budget: ${remaining_budget:.2f}")

            if budget_used_pct > 100:
                print("⚠️ You are over budget!")
            elif budget_used_pct > 80:
                print("🔶 Getting close to your budget limit.")
            else:
                print("✅ You are well within your budget.")
        else:
            print("Monthly budget not set or is invalid.")

    elif choice == "5":
        running = False
        print(f"\nGoodbye {name}!")
        print(f"Final summary: {len(expenses)} expenses, ${sum(expenses) if expenses else 0:.2f} total")
        print("🚀 Your finance tracker is growing stronger each week!")

    else:
        print("Invalid choice! Please choose 1-5.")

print("\n=== Finance Tracker Session Complete ===")
```
